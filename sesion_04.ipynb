**Conjuntos** `set`, son equivalentes a los conjuntos matemáticos, están definidos por valores separados por comas y contenidos entre llaves. Los elementos del conjunto no tienen un orden determinado y los valores duplicados son ignorados. No son secuencias y no están indizados. Están ideados para responder a la pregunta de si un elemento les pertenece o no.

```python
y={1.0, 10, "cien", (1,0,0),  1.0, 10, "pi", 3.1416, 13,5,"M", "a"}
```

```python
y
```

una lista se puede convertir en conjunto usando la función `set()`

```python
fib_conjunto=set(fib)
fib_conjunto
```

también se puede convertir una cadena de caracteres a conjunto

```python
mc_conjunto=set("Maria Skłodowska-Curie")
mc_conjunto
```

Operaciones de conjuntos:


*   Unión `|` (conjunto resultante de coleccionar los elementos de los dos conjuntos)
*   Intersección `&` (elementos comunes en los dos conjuntos)
*   Diferencia `-` (elementos en el primer conjunto pero no en el segundo)
*   Diferencia simétrica `^` (elementos que solo están en uno de los dos conjuntos pero no en ambos)
*   Subconjunto estricto `<` (`True` si el primer conjunto es subconjunto del segundo)
*   Subconjunto `<=` (`True` si el primer conjunto es subconjunto del segundo o si son iguales)
*   Superconjunto estricto `>` (`True` si el segundo conjunto es subconjunto del primero)
*   Superconjunto `>=` (`True` si el segundo conjunto es subconjunto del primero o si son iguales)



```python
y|fib_conjunto # Unión
```

```python
y&fib_conjunto # Intersección
```

```python
y-fib_conjunto # diferencia
```

```python
y^fib_conjunto # diferencia simétrica
```

```python
{1, 5, 13}<y # subconjunto estricto
```

```python
y<y
```

```python
y<=y # subconjunto
```

```python
y>fib_conjunto # superconjunto estricto
```

```python
fib_conjunto>=fib_conjunto # superconjunto
```

```python
y.discard('cien')
```

```python
y
```

```python
y.add('doscientos')
```

```python
y
```

**Diccionarios** `dict` son las estructuras de datos más importantes de Python. Son colecciones no ordenadas, mutables de pares llave/valor. Sintais: los diccionarios se escriben entre llaves `{}`, los pares llave/valor son separados por coma `(,)`, mientras que en cada par llave/valor también conocido como *item* la llave se separa del valor por dos puntos `(:)`.

```python
marie_curie={"nombre": "Marie", "apellido": "Skłodowska-Curie","fecha_nacimiento":[7,"Noviembre",1867],"premio_nobel":[1903,1911]}
```

```python
marie_curie
```

No están indizados como las listas o las tuplas, sino que  se puede acceder a cada valor almacenado en el diccionario a través de su llave

```python
marie_curie['nombre']
```

```python
marie_curie['apellido']
```

```python
marie_curie['fecha_nacimiento']
```

Los diccionarios se pueden separar en múltiples líneas

```python
constantes={
'pi': 3.14159,
'e': 2.718,
'h': 6.626e-34
}
```

```python
constantes['pi']
```

Una forma alternativa de formar diccionarios es usando una lista de tuplas de la forma (llave,valor) usndo la función `dict()`

```python
cuadrados=dict([(1,1),(2,4),(3,9),(4,16)])
```

```python
cuadrados[4]
```

Los diccionarios son mutables. Un método para modificarlos es `.update()`

```python
cuadrados.update({5:25})
```

```python
cuadrados
```

```python
cuadrados.update({2:"cuatro"})
```
