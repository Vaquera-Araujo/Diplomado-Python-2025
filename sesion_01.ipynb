**Variables en Python**

Las variables son objetos en memoria utilizados para almacenar información que puede ser referenciada o manipulada en un progama. Python utiliza un tipado dinámico, donde el tipo de variable se determina automáticamente al asignarle un valor.

Ejemplos:

```python
numero=10
flotante=12.2784356710
complejo=1+3j
cadena="Hola a todas y todos"
booleano=False
```

Python ofrece tres tipos de datos:


*   Numéricos: `int`, `float` y `complex`
*   Cadenas de caracteres: `str`
*   Booleanos: `bool`, que toman valores `True` y `False`

Podemos identificar el tipo de variable usando la función `type()`

```python
type(numero)
```

```python
type(flotante)
```

```python
type(complejo)
```

```python
type(cadena)
```

```python
type(booleano)
```

En python existen dos formas de desplegar información:

1.   A través de la evaluación de celdas
2.   Usando impresiones con la función `print()`



```python
print(type(numero),type(flotante),type(complejo))
```

La función `print()` es más versatil pues nos permite adoptar múltiples formatos, en particular f-strings como en el siguiente ejemplo:

```python
print(f"la variable numero con valor {numero}, es del tipo {type(numero)}")
```

**Operadores en Python**

Operadores Aritméticos

*   Suma `+`
*   Resta `-`
*   Multiplicación `*`
*   División `/`
*   Módulo `%`
*   Exponente `**`
*   Cociente `//`

Ejemplos:


```python
# Definir las variables x & y
x, y= 10, 3
```

```python
print("Suma:",x+y)
print("Resta:",x-y)
print("Multiplicación:",x*y)
print("División:",x/y)
print("Módulo:",x%y)
print("Exponente:",x**y)
print("Cociente:",x//y)
```
