**Operadores**

Operadores relacionales

Comparan los valores contenidos en dos variables y determinan si una relación es verdadera o falsa:


*   Igual `==`
*   No igual `!=`
*   Mayor que `>`
*   Menor que `<`
*   Mayor o igual que `>=`
*   Menor o igual que `<=`

Ejemplos:


```python
x,y= 10,3
```

```python
print("Las variables son iguales:",x==y)
print("Las variables son diferentes:",x!=y)
print("La primera variable es mayor que la segunda:",x>y)
print("La primera variable es menor que la segunda:",x<y)
print("La primera variable es mayor o igual que la segunda:",x>=y)
print("La primera variable es menor o igual que la segunda:",x<=y)
```

Operadores Lógicos

Combinan operaciones Booleanas


*   `and` devuelve `True` si ambos operandos son verdaderos
*   `or`  devuelve `True` si al menos uno de los operandos es verdadero
*   `not` invierte el valor del operando

Ejemplos

```python
verdadero, falso= True, False
```

```python
print("verdadero y verdadero:", verdadero and verdadero)
print("verdadero y falso:", verdadero and falso)
print("falso y falso:", falso and falso)
```

```python
print("verdadero y verdadero:", verdadero and verdadero)
print("verdadero y falso:", verdadero and falso)
print("falso y falso:", falso and falso)
```

```python
print("no verdadero:", not verdadero)
print("no falso:", not falso)
print("no no verdadero:", not not verdadero)
print("no no falso:", not not falso)
```

Operadores de Asignación

Modifican el valor de una variable



*   Asignación simple `=`
*   Asigna la suma del valor anterior y el valor nuevo `+=`
*   Asigna la resta del valor anterior y el valor nuevo `-=`
*   Asigna el producto del valor anterior y el valor nuevo `*=`
*   Asigna la razón del valor anterior y el valor nuevo `/=`
*   Asigna el módulo del valor anterior y el valor nuevo `%=`
*   Asigna el valor anterior elevado a la potencia del valor nuevo `**=`
*   Asigna el cociente del valor anterior con respecto al valor nuevo `//=`

Ejemplo:




```python
# Operadores de Asignación.
x,y= 10,3
#En este punto el valor de la variable x es 10
print("valor original de x:",x)
# podemos incrementar este valor
x+=y
print("x incrementado en y unidades",x)
# podemos reducir este valor
x-=y
print("x reducido en y unidades",x)
# podemos multiplicar este valor
x*=y
print("x multiplicado por y",x)
# podemos dividir este valor
x/=y
print("x dividido por y",x)
# podemos exponenciar este valor
x**=y
print("x exponenciado a la y potencia",x)
# podemos tomar el cociente este valor
x//=y
print("cociente de x con respecto a y",x)
# podemos tomar el módulo este valor
x%=y
print("módulo de x con respecto a y",x)
```

El operador Walrus `:=` se usa para hacer asignaciones de variables dentro de expresiones

```python
# Operador Walrus
x=10
print(z:=x+2)
# 'z' ahora contiene el resultado de la suma, y 'x' conserva su valor original.

# Equivalente a
z=x+2
print(z)
```

**Operaciones con Cadenas de Caracteres (Strings)**

```python
nombre="proton"
```

Indización: obtener y manipular partes de la cadena de caracteres a partir de los índices conenidos en la misma

```python
nombre[2]
```

```python
nombre[-1]
```

En python se entiende que cada uno de los caracteres en la cadena está contenido en una red definida por los valores del índice:
en este ejemplo

0p1r2o3t4o5n6

Podemos extraer rangos de caracteres usando el símbolo `:`

```python
nombre[2:5]
```

En la extracción de rangos Python se detiene en el valor final, no incluye el valor a la derecha de este. Con esta convención, la diferencia entre el valor final y el inicial es el número de letras que arroja

```python
nombre[2:]
```
